# Predicting using LinearRegression, the Factors Influencing Population Growth

```python
import pandas as pd
df=pd.read_csv("s-e.csv")
df=df.drop("MobileSubs/100",axis=1)
df=df.drop("GNIAtlas",axis=1)
df=df.drop("AdolFertRate",axis=1)
df.head(1)
```

```python
df.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
lr=LinearRegression()
x=df[['GDP','AgriValAdd%GDP','GDPGrowth%','Exports%GDP','NetMigr','FDINetBoP','SurfAreaSqKm','GNI/CapAtlas','Imports%GDP','IndValAdd%GDP','InflConsPric%','MerchTrade%GDP','PopTotal','PopDens','FertRate','LifeExpBirth','MortRateU5','UrbanPopGrowth%']]
y=df['PopGrowth%']

xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2,random_state=0)

print(xtr)
print('------------')
print(ytr)
```

```python
lr.fit(xtr,ytr)
```

```python
ypred=lr.predict(xte)
print(ypred)
```

```python
mse=mean_squared_error(yte,ypred)
print(mse)
```

```python
r2s=r2_score(yte,ypred)
print(r2s)
```
